```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sin_wave(x, w, A):
    return A * np.sin(w * x)

x = np.linspace(0, 2 * np.pi, 1000)
amplitude = np.array([0.5, 1, 1.5, 2, 2.5])
frequency = np.array([1, 2, 3, 4, 5])

fig, axs = plt.subplots(1, 3, figsize = (25, 5))

for freq, amp in zip(frequency, amplitude):
    print(f"{amp} has a corresponding frequency of {freq}")

for amp in amplitude:
    y = make_sin_wave(x = x, w = 1, A = amp)
    axs[0].plot(x, y, label = f"amplitude - {amp}")
axs[0].set_title("Varying x-values")
axs[0].set_xlabel("x-value")
axs[0].set_ylabel("y-value")
axs[0].legend()

for freq in frequency:
    y = make_sin_wave(x = x, w = freq, A = 1)
    axs[1].plot(x, y, label = f"frequency - {freq}")
axs[1].set_title("Varying frequency")
axs[1].set_xlabel("x-value")
axs[1].set_ylabel("y-value")
axs[1].legend()

for amp, freq in zip(amplitude, frequency):
    y = make_sin_wave(x = x, w = freq, A = amp)
    axs[2].plot(x, y, label = f"frequency - {freq}, amplitude - {amp}")
    
axs[2].set_title("Varying amplitude")
axs[2].set_xlabel("x-value")
axs[2].set_ylabel("y-value")
axs[2].legend()



plt.show
```
