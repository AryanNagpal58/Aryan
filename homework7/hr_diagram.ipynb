```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows = 1)
```

```python
print(data.shape)
# 3.2 - Analyzing Imported Data

# Steps:

# 1. When printing the shape of the 2D array 'data', it is shown to have 2642 rows.

# 2. 1rst column - magnitude, 2nd column - temperature
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]

highest_mag = np.max(magnitude)
lowest_mag = np.min(magnitude)

highest_temp = np.max(temperature)
lowest_temp = np.min(temperature)

#4.
print(highest_mag) #Corresponding Star: WASP-61
print(lowest_mag) #Corresponding Star: Schedar
#5.
print(highest_temp) #Corresponding Star: B-type or A-type
print(lowest_temp) #Corresponding Star: K-type

high_mag_ind = np.argmax(magnitude)
low_mag_ind = np.argmin(magnitude)
high_temp_ind = np.argmax(temperature)
low_temp_ind = np.argmin(temperature)

stars = np.array([
    data[high_mag_ind],
    data[low_mag_ind],
    data[high_temp_ind],
    data[low_temp_ind]
])


# 6. Recall: The star WASP-61 is the dimmest in the data, and the star Schedar is the brightest of those in the data
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

plt.figure(figsize = (8, 6))
plt.scatter(log_temperature, log_magnitude, color = 'gray', alpha = 0.4, s = 50, label = "All stars plotted")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

xstars = np.log(stars[:, 1])
ystars = np.log(stars[:, 0])

color_coding = ['purple', 'black', 'green', 'orange']
labels = ["Dimmest star", "Brightest Star", "Hottest Star", "Coldest Star"]

for i in range(len(stars)):
    plt.scatter(xstars[i], ystars[i], color = color_coding[i], alpha = 0.6, s = 150, label = labels[i])

plt.xlabel("Temperature (Log K)")
plt.ylabel("Magnitude")
plt.title("Hertzsprung-Russel Diagram")
plt.legend()
plt.show()
```
