```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.table import Table
from scipy.optimize import curve_fit

table = Table.read('../../Aryan Nagpal/homework11/global_CCl4_MM.dat', format = 'ascii.basic')

df = table.to_pandas()

df = df[['CCl4ottoyr', 'CCl4ottoGLm', 'CCl4ottoGLsd']]
df = df.rename(columns = {'CCl4ottoyr' : 'date', 'CCl4ottoGLm' : 'global mean concentration', 'CCl4ottoGLsd' : 'global mean concentration sd'})

x = df['date']
y = df['global mean concentration']
y_error = df['global mean concentration sd']

mask = np.isfinite(x) & np.isfinite(y) & np.isfinite(y_error)
x_clean = x[mask]
y_clean = y[mask]
y_error_clean = y_error[mask]

def model (x, A, k):
    return A - (k*x)

p0 = [y_clean.max(), 0.1]
popt, pcov = curve_fit(model, x_clean, y_clean, sigma = y_error_clean, absolute_sigma = True, p0 = p0)
A_fit, k_fit = popt

x_fit = np.linspace(x_clean.min(), x_clean.max(), 100)
y_fit = model(x_fit, A_fit, k_fit)

plt.figure(figsize = (8,6))

plt.errorbar(
    x,
    y,
    yerr = y_error,
    fmt = 'o',
    ecolor = 'red',
    capsize = 5,
    label = "Mean Concentration"
    
)
plt.plot(x_fit, y_fit, color = 'red', label = 'Line of fit')
plt.xlabel("Data")
plt.ylabel("Mean Concentration")
plt.title("Global Mean Concentration Including Error")
plt.legend()
plt.show()

residuals = y_clean - model(x, A_fit, k_fit)
chi2 = np.sum(residuals**2)
dof = len(x) - len(popt)
chi2_red = chi2 / dof

print(f"Reduced Chi-Sqared for Global Concentration Data: {chi2_red:.2f}")
print("The model above appears to fit well with the plotted data of global mean concentration vs. date, with still some obvious variance or error observable.")
```
