```python
import pandas as pd
df = pd.read_csv("state_data.csv")
# sd = pd.read_csv("../course_assignments/homework8/state_data/csv")
print(df.shape) #4. This shape has 4 columns and 52 rows
print(df.head())
print(df.tail())
# 5.
# Each column reprsents, respectively, the first two letters of a US state, the state's corresponding latitude, longitude, and then full name.
# The first state in the data set is Alaska
# The last state in the data set is Wynoming
# There are 52 x 4 entries, or 208 entries overall
# There are more than 50 rows because there is a row for labels for each column, and a row for the district of columbia. These two rows, plus the 50 states,
# make up the 52 states of the data set

med_lat = df['latitude'].median()
med_long = df['longitude'].median()
print(med_lat)
print(med_long)

df['abs. from lat'] = abs(df['latitude'] - med_lat)

df['abs. from long'] = abs(df['longitude'] - med_long)
print(df.shape) # 9. There are now 52 rows and 6 columns

closest_med_lat = df['abs. from lat'].min()
closest_med_lat_index = df['abs. from lat'].argmin()
closest_med_lat_name = df.iloc[closest_med_lat_index]['name']

far_med_lat = df['abs. from lat'].max()
far_med_lat_index = df['abs. from lat'].argmax()
far_med_lat_name = df.iloc[far_med_lat_index]['name']


print(f"The closest to the median of latitude is {closest_med_lat_name} by {closest_med_lat:.2f}")
print(f"The farthest from the median of latitude is {far_med_lat_name} by {far_med_lat:.2f}")


closest_med_long = df['abs. from long'].min()
closest_med_long_index = df['abs. from long'].argmin()
closest_med_long_name = df.iloc[closest_med_long_index]['name']

far_med_long = df['abs. from long'].max()
far_med_long_index = df['abs. from long'].argmax()
far_med_long_name = df.iloc[far_med_long_index]['name']

print(f"The closest to the median of longitude is {closest_med_long_name} by {closest_med_long:.2f}")
print(f"The farthest from the median of longitude is {far_med_long_name} by {far_med_long:.2f}")

#11. Pandas function from website: .sort_values()

sorted_by_lat = df.sort_values(by='latitude', ascending = False)
#This function was found under the 'Viewing Data' Section of the subsection '10 Minutes to Pandas' of the Pandas User Guide on the documentation website
#This function sorts the rows of a data frame based on the order of value for a specific column. That column is defined as a parameter within the brackets of .sort_values().
#There are more possible parameters that can be added for describing the newly ordered version of the data frame, such as whether the values are in descending or ascending order.

print(sorted_by_lat)
```

```python
# 3 - Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

sdf = sns.load_dataset('diamonds')

print(sdf)

sdf_no_NaN = sdf.dropna()
# 5. I know that the columns of cut, color and clarity should have been taken away with this, but for some reason they would not. I just did not use them later on.

# 6. Answer the following by analyzing your chosen data set.

print(sdf_no_NaN.shape)
# The shape of the data set 'diamonds' seems to be 53940x10
# There are 53,940 rows, and 10 columns

#Printing the first five and last five columns:
print(sdf_no_NaN.head())
print(sdf_no_NaN.tail())

# Names of each column:
# - carat - The carat weight of the diamond described for said row
# - cut - The grade of the diamon described for said row, in terms of proportional symmetry and angles
# - color - The degree of the tint of the diamond for said row
# - clarity - The grade of clarity of the diamond for said row
# - depth - Measurement of a diamond's height, from the top to the bottom. Measured usually in millimeters
# - table - Measurement of the largest, flattest, horizontal surface of a diamond.
# - price - The price given for the diamond, expected to be in USD
# x - The x-dimensional length of the diamond. Unspecified what the unit of measurement is, but assuming millimeters.
# y - The y-dimensional length of the diamond. Unspecified what the unit of measurement is, but assuming millimeters.
# z - The z-dimensional length of the diamond. Unspecified what the unit of measurement is, but assuming millimeters.


fig, axs = plt.subplots(2, 2)
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
carat = sdf_no_NaN['carat']
depth = sdf_no_NaN['depth']
table = sdf_no_NaN['table']
price = sdf_no_NaN['price']
x_meas = sdf_no_NaN['x']
y_meas = sdf_no_NaN['y']
z_meas = sdf_no_NaN['z']

axs[0,0].scatter(x = carat, y= price, color = 'red')
axs[0,0].set_title("Carat vs. Price")
axs[0,0].set_xlabel("Carat Grade by Unit Carat")
axs[0,0].set_ylabel("Price of Diamond (USD)")

# This subplot represented the relationship between each diamond's carat value and its overall price.
# I chose these two columns because I wanted to test the possibility of a relationship between two commonly related traits of diamonds: their carat and their price.
# Through this subplot, I was able to find a directly proportional relationship between the carat measurement of diamonds in the data set and each diamond's overall price

axs[0,1].scatter(x = x_meas, y = y_meas, color = 'blue')
axs[0,1].set_title("X-Dimension vs. Y-Dimension Scatter Plot")
axs[0,1].set_xlabel("Length (mm)")
axs[0,1].set_ylabel("Width (mm)")
#

sns.histplot(data = sdf_no_NaN, x = 'depth', ax=axs[1,0], color = 'purple')
axs[1,0].set_title("Depth by Histogram")
axs[1,0].set_xlabel("Depth from table to culet (mm)")
#
sns.boxplot(data = sdf_no_NaN, x = 'table', ax=axs[1,1], color = 'yellow', label = 'Inner Quartile Range, Q1-Q3 for table (mm)')
axs[1,1].set_title("table length by boxplot")
axs[1,1].set_xlabel("table length (mm)")
axs[1,1].legend()
#

plt.show()



```
