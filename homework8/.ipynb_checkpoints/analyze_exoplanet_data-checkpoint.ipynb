```python
#4.2 Importing Data
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('exoplanet_data.csv', comment = "#")

print(df.shape)
# 4. The shape of this data set is 576x92. There are 576 rows, and 92 columns.
print(df.columns)
# 5 column titles that I have no idea about would include 'pl_insolerr1', 'pl_isollim', 'pl_eqterr1', 'pl_eqtlim', and 'ttv_flag'

```

```python
# 4.3 Reducing Data

# Table Label - Planet Name, Database Column Name - 'pl_name',
# Descrip:
# Table Label - Host Name, Database Column Name - 'hostname',
# Descrip:
# Table Label - Discovery Method, Database Column Name - 'discoverymethod',
# Descrip:
# Table Label - Spectral Type, Database Column Name - 'pl_radj',
# Descrip:
# Table Label - Equilibrium Temperature (K), Database Column Name - 'pl_radeerr1',
# Descrip:
# Table Label - Planet Mass or Mass*sin(i) [Jupiter Mass], Database Column Name - 'pl_orbsmaxerr2',
# Descrip:

dcn = df[['pl_name','hostname','discoverymethod','pl_radj','pl_radeerr1','pl_orbsmaxerr2']]

dcn_only_num = dcn.dropna()

print(dcn_only_num.shape)

#Shape of the smaller data set: 133x6. There are 133 rows and 6 columns

print(dcn_only_num.head())
print(dcn_only_num.tail())

print(dcn_only_num.columns)
```

```python
#4.4 Analyze Data

print(dcn_only_num['discoverymethod'].value_counts())

# The different discovery methods include transit, radial velocity, and imaging
# There are 124 for transit, the most, 8 for radial velocity, and for the least, only 1 for imaging
# In my own words:
# Transit: A form of discovering an exoplanet by finding a dip in the brightness of an observed star, caused by the discovered exoplanet getting in between the star and the observer.
# Radial Velocity: By studying the small periodic shifts in the wavelength of nearby stars, exoplanets that are nearby and causing these shifts with their graviational pull can be discovered.
# Imaging: Discovering exoplanets by taking direct photos of them by blocking the light of the stars that host them.


color_method = dcn_only_num['discoverymethod']

plt.figure(figsize = (8,6))

color1 = ['red', 'blue', 'yellow']

for method in color_method.unique():
    sub = dcn_only_num[color_method == method]
    plt.scatter(x = sub['pl_orbsmaxerr2'], y = sub['pl_radeerr1'], label = color_method, color = color1, alpha = 0.5)

    
    

```
